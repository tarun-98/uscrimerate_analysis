#### Indexes of cities which has got the highest crime rate

```python
crime_stats=pd.DataFrame()
for i in ['murdPerPop', 'rapesPerPop','robbbPerPop','assaultPerPop','burglPerPop','larcPerPop', 'autoTheftPerPop', 'arsonsPerPop']:
    mud=[]
    for j in state:
        mud.append(np.argmax(j[1][i]))
    crime_stats[i]=mud
crime_stats.index=state_stats.index
crime_stats.head(10)
```

```python
def get_coord(crime):
    city=pd.DataFrame()
    state=[]
    lats,long,name,crime_rate=[],[],[],[]
    for i,j,k in zip(data2,data.loc[crime_stats[crime],'city'].values,list(data.loc[crime_stats[crime],crime].values)):
        if len(i[1].name[i[1].name==j])>0:
            #print(i[1].name[i[1].name==j])
            lats.extend(i[1].lat[i[1].name==j].values)
            long.extend(i[1].lng[i[1].name==j].values)
            crime_rate.append(k)
            name.append(j)
            state.append(i[0])
    city['state']=state
    city['crime_rate']= crime_rate       
    city['lats']=lats
    city['long']=long
    city['name']=name
    return(city)
```

```python
def state_plot(feat):
    data=[]
    for i in range(2):
        if (i==0):
            geo_key = 'geo'+str(i+1) if i != 0 else 'geo'
            # Walmart store data
            data.append(
                dict(type='choropleth',
            locations=state_stats.index, # Spatial coordinates
            z = state_stats[feat], # Data to be color-coded
            locationmode = 'USA-states', # set of locations match entries in `locations`
            colorscale = 'Reds',
            colorbar_title = str(feat)))
            # Year markers

            #print(data)
            lay[geo_key] = dict(
                scope = 'usa',
                showland = True,
                landcolor = 'rgb(229, 229, 229)',
                showcountries = False,
                domain = dict( x = [], y = []))
        else:
            city=get_coord(feat)
            data.append(go.Scattergeo(
            lon = city['long'],
            lat = city['lats'],
            text = city['name'],
            mode = 'markers',
            marker_color = city['crime_rate'],
            ))


    z = 0
    COLS = 2
    ROWS = 1
    for x in range(ROWS):
        for y in range(COLS):
           # print(geo_key)
            geo_key = 'geo'+str(z+1) if z != 0 else 'geo'
            lay[geo_key]['domain']['y'] = [float(x)/float(ROWS), float(x+1)/float(ROWS)]
            lay[geo_key]['domain']['x'] = [float(y)/float(COLS), float(y+1)/float(COLS)]
            z=z+1



    for i in range(2):
        fig = go.Figure(data=data[i], layout=lay)
        fig.show()
```
